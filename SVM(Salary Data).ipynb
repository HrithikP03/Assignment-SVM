```python
string_columns = ['workclass', 'education', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex', 'native', 'Salary']
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for i in string_columns:
    traindata[i] = LE.fit_transform(traindata[i])
    testdata[i] = LE.fit_transform(testdata[i])
```

```python
traindata
```

```python
testdata
```

```python
x_train = traindata.iloc[0:500,0:13]
y_train = traindata.iloc[0:500,13]
x_test = testdata.iloc[0:300,0:13]
y_test = testdata.iloc[0:300,13]
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
model_linear = SVC(kernel = 'linear')
model_linear.fit(x_train, y_train)
train_pred_lin = model_linear.predict(x_train)
test_pred_lin = model_linear.predict(x_test)
train_lin_acc = np.mean(train_pred_lin == y_train)
test_lin_acc = np.mean(test_pred_lin == y_test)
print("Accuracy of train data using linear:",train_lin_acc)
print("Accuracy of test data using linear:",test_lin_acc)
```

```python
model_poly = SVC(kernel = 'poly')
model_poly.fit(x_train, y_train)
train_pred_pol = model_poly.predict(x_train)
test_pred_pol = model_poly.predict(x_test)
train_pol_acc = np.mean(train_pred_pol == y_train)
test_pol_acc = np.mean(test_pred_pol == y_test)
print("Accuracy of train data using poly:",train_pol_acc)
print("Accuracy of test data using poly:",test_pol_acc)
```

```python
model_rbf = SVC(kernel = 'rbf')
model_rbf.fit(x_train,y_train)
train_pred_rbf = model_rbf.predict(x_train)
test_pred_rbf = model_rbf.predict(x_test)
train_rbf_acc = np.mean(train_pred_rbf == y_train)
test_rbf_acc = np.mean(test_pred_rbf == y_test)
print("Accuracy of train data using RBF:",train_rbf_acc)
print("Accuracy of test data using RBF:", test_rbf_acc)
```
